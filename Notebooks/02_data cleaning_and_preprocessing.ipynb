Extracing and cleaning key datapoints

```python
def clean_data(df):
    def safe_extract(data, key, separator="|"):
        if pd.isna(data) or data == '':
            return ''
        try:
            if isinstance(data, str):
                data = eval(data)
            if isinstance(data, list):
                return separator.join([item.get(key, '') for item in data])
            elif isinstance(data, dict):
                return data.get(key, '')
            else:
                return str(data)
        except (SyntaxError, TypeError, AttributeError, NameError):
            return ''

    df['belongs_to_collection'] = df['belongs_to_collection'].apply(lambda x: safe_extract(x, 'name'))
    df['genres'] = df['genres'].apply(lambda x: safe_extract(x, 'name', separator="|"))
    df['spoken_languages'] = df['spoken_languages'].apply(lambda x: safe_extract(x, 'name', separator="|"))
    df['production_countries'] = df['production_countries'].apply(lambda x: safe_extract(x, 'name', separator="|"))
    df['production_companies'] = df['production_companies'].apply(lambda x: safe_extract(x, 'name', separator="|"))

    return df

df = clean_data(df)
print(df[['belongs_to_collection', 'genres', 'spoken_languages', 'production_countries', 'production_companies']].head())
```

Replacing Unrealistic values

* Replacing '0' with NaN in Budget, Revenue, and Runtime.

```python
import numpy as np
def replace_zero_with_nan(df, columns):
    for col in columns:
        df[col] = df[col].replace(0, np.nan)
    return df

columns_to_check = ['budget', 'revenue', 'runtime']
df = replace_zero_with_nan(df, columns_to_check)
```

* Inferring values from similar movies to fill NaNs using median

```python
for col in columns_to_check:
  df[col] = df.groupby('genres')[col].transform(lambda x: x.fillna(x.median()))
  df[col] = df[col].fillna(df[col].median())
```

**Reordering columns**

Some of the colums requested are not part of the original movies dataset, so it has to be extracted from the same API but fetching 'credits' data.

```python
import requests
import pandas as pd

API_KEY = "N/A"
MOVIE_IDS = [0, 299534, 19995, 140607, 299536, 597, 135397,
             420818, 24428, 168259, 99861, 284054, 12445,
             181808, 330457, 351286, 109445, 321612, 260513]

def fetch_movie_details(movie_id):
    url = f"https://api.themoviedb.org/3/movie/{movie_id}/credits"
    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "accept": "application/json"
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.HTTPError as e:
        print(f"Error fetching credits data {movie_id}: {e}")
        return None
    except Exception as e:
        print(f"Unexpected error for credits data {movie_id}: {e}")
        return None

credits_data = []
for movie_id in MOVIE_IDS:
    movie = fetch_movie_details(movie_id)
    if movie:
        credits_data.append(movie)
    else:
        print(f"Skipped movie ID {movie_id}")

data = pd.DataFrame(credits_data)

if 'genres' in data.columns:
    data['genres'] = data['genres'].apply(lambda x: ', '.join([g['name'] for g in x]) if isinstance(x, list) else '')
else:

# Creating an empty 'genres' column to handle the case where 'genres' column is missing
    data['genres'] = ''

print(f"\nSuccessfully fetched {len(data)} movies")
print("\nDataFrame info:")
print(data.info())
print("\nFirst 5 movies:")
print(data.head())

# Saving the Credits dataset in CSV format
data.to_csv('credits.csv', index=False)
print("\nData saved to 'credits.csv'")
```

* Creating cast-size column based on cast column

```python
import pandas as pd

def get_cast_size(cast_data):
    try:

        if isinstance(cast_data, str) and cast_data.startswith('[') and cast_data.endswith(']'):
            cast_data = eval(cast_data)

        if isinstance(cast_data, list):
            return len(cast_data)

        return 0

    except (SyntaxError, TypeError, NameError):
        return 0

credits['cast_size'] = credits['cast'].apply(get_cast_size)
print(credits[['cast', 'cast_size']].head())
```

```python
credits.head(5)
```

* Creating crew_size column based on crew column.

```python
def get_crew_size(crew_data):
    try:
        if isinstance(crew_data, str) and crew_data.startswith('[') and crew_data.endswith(']'):
            crew_data = eval(crew_data)
        if isinstance(crew_data, list):
            return len(crew_data)
        return 0
    except (SyntaxError, TypeError, NameError):
        return 0

credits['crew_size'] = credits['crew'].apply(get_crew_size)
print(credits[['crew', 'crew_size']].head())
credits.head()
```

```python
credits['director'] = credits['crew'].apply(
    lambda x: next((crew['name'] for crew in x if crew['job'] == 'Director'), None)
)
```
